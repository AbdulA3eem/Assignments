```python
# using for statement

paper_thickness_M = 0.00008 

folded_thickness_M = paper_thickness_M

num_folds = 43

for _ in range(num_folds):
    folded_thickness_M = folded_thickness_M * 2

thickness_10000_km = folded_thickness_M / 1000

print(f'The thickness of paper folded {num_folds} times is {thickness_10000_km:.2f}','KM')
```

```python
# saving to list

paper_thickness_meters = 0.00000008 # in km 

thickness_history_meters = []

thickness_history_meters.append(paper_thickness_meters)

num_folds = 43

for _ in range(num_folds):
    folded_thickness_meters = thickness_history_meters[-1] * 2
    thickness_history_meters.append(folded_thickness_meters)
```
